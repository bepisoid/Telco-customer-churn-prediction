# Business Value Analysis

## Model Performance Summary (Holdout Set)
| Metric | Value | Interpretation |
|--------|-------|----------------|
| **AUC** | 0.837 | Strong discriminatory power between churners and non-churners. |
| **Recall** | 0.793 | Captures **79.3%** of all potential churners. |
| **Precision** | 0.506 | ~50% of targeted customers are actual churners. |
| **F1-Score** | 0.618 | Balanced harmonic mean of Precision and Recall. |

## Financial Impact Simulation

### Assumptions
- **Customer Base:** 10,000 active users.
- **Base Churn Rate:** 27% (approx. 2,700 at-risk customers).
- **Customer Lifetime Value (LTV):** $600.
- **Retention Campaign Cost:** $50 per targeted customer (incentives/marketing).
- **Retention Success Rate:** 50% (probability of saving a customer if targeted).

### Projected P&L (Profit and Loss)

1.  **Campaign Reach (Who we target):**
    Based on the model's Precision (50.6%) and Recall (79.3%), to reach **2,141** true churners, we must target approximately **4,231** customers (including false positives).

2.  **Costs:**
    *   Total Targeted Users × Cost per User
    *   4,231 × $50 = **$211,550**

3.  **Revenue Retained:**
    *   True Churners Found × Success Rate × LTV
    *   2,141 × 50% × $600 = **$642,300**

4.  **Net Profit:**
    *   $642,300 - $211,550 = **$430,750**

### Return on Investment (ROI)
The model generates a projected **203% ROI** on the retention budget.

> By proactively targeting the top 42% most risky customers identified by the model, the company can save over **$430k annually** compared to a reactive approach.
